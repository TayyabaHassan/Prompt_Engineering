### Creating Parameter Configurations

```python
def get_completion(prompt,
                   model="llama-3.1-8b-instant",
                   temperature=0,
                   top_p=1,
                   max_tokens=200,
                   frequency_penalty=0,
                   presence_penalty=0):

    response = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "user", "content": prompt}
        ],
        temperature=temperature,
        top_p=top_p,
        max_tokens=max_tokens,
        frequency_penalty=frequency_penalty,
        presence_penalty=presence_penalty
    )

    return response.choices[0].message.content
```

```python
prompt = "Write a 150-word explanation of how artificial intelligence is used in healthcare."
```

### Running the Experiments

```python
experiments = [
    {"temp":0.1, "top_p":1, "max_tokens":200, "freq":0, "pres":0},
    {"temp":0.5, "top_p":1, "max_tokens":200, "freq":0, "pres":0},
    {"temp":0.9, "top_p":1, "max_tokens":200, "freq":0, "pres":0},
    {"temp":0.7, "top_p":0.8, "max_tokens":200, "freq":0, "pres":0},
    {"temp":0.7, "top_p":0.5, "max_tokens":200, "freq":0, "pres":0},
    {"temp":0.7, "top_p":1, "max_tokens":100, "freq":0, "pres":0},
    {"temp":0.7, "top_p":1, "max_tokens":200, "freq":1, "pres":0},
    {"temp":0.7, "top_p":1, "max_tokens":200, "freq":0, "pres":1}
]
```

### REsults

```python
results = []

for i, exp in enumerate(experiments):

    output = get_completion(
        prompt,
        temperature=exp["temp"],
        top_p=exp["top_p"],
        max_tokens=exp["max_tokens"],
        frequency_penalty=exp["freq"],
        presence_penalty=exp["pres"]
    )

    results.append({
        "Experiment": i+1,
        "Temperature": exp["temp"],
        "Top_p": exp["top_p"],
        "Max_tokens": exp["max_tokens"],
        "Frequency_penalty": exp["freq"],
        "Presence_penalty": exp["pres"],
        "Output": output
    })
```

# Part 4: Multi-Step Prompt Pipeline

### 4.1 Input News Text

```python
news_text = """
Artificial intelligence is rapidly transforming healthcare. Hospitals are using AI systems
to detect diseases earlier, analyze medical images, and assist doctors in diagnosis.
AI-powered tools can process large volumes of patient data and identify patterns that
humans might miss. Researchers believe AI will improve treatment accuracy and reduce
healthcare costs in the future.
"""
```

### 4.2 Step 1: Summarizing the Article

```python
summary_prompt = f"""
Summarize the following news article in 3 sentences:

{news_text}
"""

summary_output = get_completion(summary_prompt)

print(summary_output)
```

### 4.3 Step 2: Extracting Key Facts


```python
facts_prompt = f"""
From the following summary extract 3 key facts:

{summary_output}
"""

facts_output = get_completion(facts_prompt)

print(facts_output)
```

### 4.4 Step 3: Topic Classification


```python
classification_prompt = f"""
Classify the topic of the following information into one category:
Politics, Technology, Health, Business, or Environment.

Text:
{facts_output}
"""

classification_output = get_completion(classification_prompt)

print(classification_output)
```

### 4.5 Step 4: Tweet-Length Summary

```python
tweet_prompt = f"""
Write a tweet-length summary (max 30 words) based on the following facts:

{facts_output}
"""

tweet_output = get_completion(tweet_prompt)

print(tweet_output)
```

### 4.6 Pipeline Results

```python
pipeline_results = {
    "Stage": [
        "Summary",
        "Key Facts",
        "Topic Classification",
        "Tweet Summary"
    ],
    "Output": [
        summary_output,
        facts_output,
        classification_output,
        tweet_output
    ]
}

df_pipeline = pd.DataFrame(pipeline_results)

df_pipeline
```
